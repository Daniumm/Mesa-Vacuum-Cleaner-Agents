Igualmente podemos ver en algunos casos líneas horizontales que significa que por varios pasos del modelo no hubo una reducción del numero de bloques sucios, lo que nos da a entender que las aspiradoras en si se enfocan en ciertos puntos que ya estan limpios y no recogen las otras partes en las que se deberían enfocar.

Los agentes, que en este caso son las aspiradoras no son racionales, ya que no tienen conocimiento del ambiente, sus movimientos son aleatorios por lo que no elige el mejor camino todas las veces.

El ambiente es observable ya que podemos ver todo al mismo tiempo, estocástico ya que no es determinista, en un punto de la simulación, no puedes saber cual será el siguiente estado, sino que es un tema de probabilidad, gracias a la parte donde se mueven aleatoriamente, también es discreto ya que es por pasos.
